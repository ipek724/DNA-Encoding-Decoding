```python
from colorama import Fore, Style

def calcRedundantBits(m):
 
    # Use the formula 2 ^ r >= m + r + 1
    # to calculate the no of redundant bits.
    # Iterate over 0 .. m and return the value
    # that satisfies the equation
 
    for i in range(m):
        if(2**i > m + i + 1):
            return i
 
def posRedundantBits(data, r):
 
    # Redundancy bits are placed at the positions
    # which correspond to the power of 2.
    j = 0
    k = 1
    m = len(data)
    res = ''
 
    # If position is power of 2 then insert '0'
    # Else append the data
    for i in range(1, m + r+1):
        if(i == 2**j):
            res = res + '0'
            j += 1
        else:
            res = res + data[-1 * k]
            k += 1
 
    # The result is reversed since positions are
    # counted backwards. (m + r+1 ... 1)
    return res[::-1]
 
 
def calcParityBits(arr, r):
    n = len(arr)
 
    # For finding rth parity bit, iterate over
    # 0 to r - 1
    for i in range(r):
        val = 0
        for j in range(1, n + 1):
 
            # If position has 1 in ith significant
            # position then Bitwise OR the array value
            # to find parity bit value.
            if(j & (2**i) == (2**i)):
                val = val ^ int(arr[-1 * j])
                # -1 * j is given since array is reversed
 
        # String Concatenation
        # (0 to n - 2^r) + parity bit + (n - 2^r + 1 to n)
        arr = arr[:n-(2**i)] + str(val) + arr[n-(2**i)+1:]
    return arr

def detectError(arr, nr):
    n = len(arr)
    res = 0
 
    # Calculate parity bits again
    for i in range(nr):
        val = 0
        for j in range(1, n + 1):
            if(j & (2**i) == (2**i)):
                val = val ^ int(arr[-1 * j])
 
        # Create a binary no by appending
        # parity bits together.
        res = res + val*(10**i)
 
    # Convert binary to decimal
    return int(str(res), 2)



'''
get string and read from file txt
'''

first = open("./encode_output.txt", "r")
firstData = first.read()
first.close()

print("First Data is : \n" + firstData + " .\n")
lenghtString = len(firstData)
print("Lenght Input String : " + str(lenghtString))

text = open("./output.txt", "r")
data = text.read()
text.close()
print("Second Data is : \n" + data + " .\n")
'''
define char with this num
'''
a = '00'
t = '01'
c = '10'
g = '11'

'''
seperator sentence with 96 char
'''
seperator = 128
lenghtString = len(data)

print("Lenght Input String : " + str(lenghtString))

listChar = []
string = ""

'''
for seperator sentence and ...
'''
for counter in range(lenghtString):
    string += data[counter]
    if ((counter+1) % seperator == 0 and counter != 0) or counter == lenghtString - 1:
        listChar.append(string)
        string = ""

print("We Found " + str(len(listChar)) + " String with lenght 128 or lower.\n")
addressString = []
counter = 0
address = ''

'''
convert to the binary 
'''
binaryString = []
string = ''
for item in listChar:
    for character in item:
        if character == 'A':
            string += '00'
        elif character == 'T':
            string += '01'
        elif character == 'C':
            string += '10'
        elif character == 'G':
            string += '11'
    binaryString.append(string)
    string = ''
    
seperator = 32
listChar = []
for item in binaryString:
    for counter in range(len(item)):
        string += item[counter]
        if ((counter+1) % seperator == 0 and counter != 0) or counter == len(item) - 1:
            listChar.append(string)
            string = ""


print()
counter = 1          
for item in binaryString:
    print("item"+ str(counter)+ " is : "+item+" with len " + str(len(item)))
    counter += 1


print("\nRead Text 2 and check it.\n")

correction = 0
counter = 0
finalstring = []


counter=0
for item in listChar:
    counter += 1
    print("item "+str(counter)+" : "+str(item) + " With size : "+str(len(item)))
    if len(item) == 32:
        correction = detectError(item, 6)
    if(correction==0):
        print("There is no error in the received message.")
        finalstring+=item
    else:
        print('Wrong Bit is '+str(len(item)-correction+1))
        item = list(item)
        print('Item shows with red Wrong bit : ')
        for count in range(len(item)):
            if count+1 == len(item)-correction+1:
                print(Fore.RED + item[count]+Style.RESET_ALL, end='')
            else:
                print(item[count], end='')
        print()
        if item[len(item)-correction] == '0' :
            item[len(item)-correction] = '1'
            finalstring+=item

        else:
            item[len(item)-correction] = '0'
            finalstring+=item

        print('Modified item with green color :')
        for count in range(len(item)):
            if count+1 == len(item)-correction+1:
                print(Fore.GREEN + item[count]+Style.RESET_ALL, end='')
            else:
                print(item[count], end='')
        print()
            
output = open("finalbinaryafter.txt", "w")
output.writelines(finalstring)
output.close()
```
