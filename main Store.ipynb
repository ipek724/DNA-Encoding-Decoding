```python
def calcRedundantBits(m):
 
    # Use the formula 2 ^ r >= m + r + 1
    # to calculate the no of redundant bits.
    # Iterate over 0 .. m and return the value
    # that satisfies the equation
 
    for i in range(m):
        if(2**i > m + i + 1):
            return i
 
def posRedundantBits(data, r):
 
    # Redundancy bits are placed at the positions
    # which correspond to the power of 2.
    j = 0
    k = 1
    m = len(data)
    res = ''
 
    # If position is power of 2 then insert '0'
    # Else append the data
    for i in range(1, m + r+1):
        if(i == 2**j):
            res = res + '0'
            j += 1
        else:
            res = res + data[-1 * k]
            k += 1
 
    # The result is reversed since positions are
    # counted backwards. (m + r+1 ... 1)
    return res[::-1]
 
 
def calcParityBits(arr, r):
    n = len(arr)
 
    # For finding rth parity bit, iterate over
    # 0 to r - 1
    for i in range(r):
        val = 0
        for j in range(1, n + 1):
 
            # If position has 1 in ith significant
            # position then Bitwise OR the array value
            # to find parity bit value.
            if(j & (2**i) == (2**i)):
                val = val ^ int(arr[-1 * j])
                # -1 * j is given since array is reversed
 
        # String Concatenation
        # (0 to n - 2^r) + parity bit + (n - 2^r + 1 to n)
        arr = arr[:n-(2**i)] + str(val) + arr[n-(2**i)+1:]
    return arr

'''
get string and read from file txt
'''
text = open("./encode_output.txt", "r")
data = text.read()
text.close()

'''
define char with this num
'''
a = '00'
t = '01'
c = '10'
g = '11'

'''
seperator sentence with 96 char
'''
seperator = 96
lenghtString = len(data)

print("Lenght Input String : " + str(lenghtString))

listChar = []
string = ""

'''
for seperator sentence and ...
'''
for counter in range(lenghtString):
    string += data[counter]
    if ((counter+1) % seperator == 0 and counter != 0) or counter == lenghtString - 1:
        listChar.append(string)
        string = ""

print("We Found " + str(len(listChar)) + " String with lenght 96 or lower.\n")
addressString = []
counter = 0
address = ''

'''
add A or T or C or G for number of a or t or c or g
'''
for item in listChar:
    binary = '{0:016b}'.format(counter)
    for count in range(len(binary)):
        string += binary[count]
        if ((count+1) % 2 == 0):
            if string == a :
                address += 'A'
            elif string == t:
                address += 'T'
            elif string == c:
                address += 'C'
            elif string == g:
                address += 'G'
            string = ''

    addressString.append(address + item)
    address = ''
    counter += 1



'''
convert to the binary 
'''

binaryString = []
string = ''
for item in addressString:
    for character in item:
        if character == 'A':
            string += '00'
        elif character == 'T':
            string += '01'
        elif character == 'C':
            string += '10'
        elif character == 'G':
            string += '11'
    binaryString.append(string)
    string = ''

counter = 1
for item in addressString:
    print("item "+ str(counter)+ " is : "+item+" with len :" + str(len(item)))
    counter += 1

counter=1
for item in binaryString:
    print("item "+ str(counter)+ " is : "+item+" with len :" + str(len(item)))
    counter += 1



seperator = 26
listChar = []
listChartext = []
for item in binaryString:
    for counter in range(len(item)):
        string += item[counter]
        if ((counter+1) % seperator == 0 and counter != 0) or counter == len(item) - 1:
            listChar.append(string)
            string = ""

binaryString = []
showString = []
count = 0
counter = 1

for counter in range(len(listChar)):
    counter += 1

    data = listChar[counter-1]
    m = len(listChar[counter-1])
    r = calcRedundantBits(m)
    # Determine the positions of Redundant Bits
    arr = posRedundantBits(data, r)
    # Determine the parity bits
    arr = calcParityBits(arr, r)
    string += arr
    showString.append(arr)
    if counter % 8 == 0 or counter == len(listChar)-1:
        binaryString.append(string)
        string = ''
print("")
counter=1
for item in binaryString:
    print("item "+ str(counter)+ " is : "+item+" with len :" + str(len(item)))
    counter += 1


checkbit = ''
counter = 0
string = ''
characterString = []

for item in showString:

    for bit in item:
        
        counter += 1
        checkbit += bit
        if counter % 2 == 0:
            if checkbit == '00':
                string += 'A'
            elif checkbit == '01':
                string += 'T'
            elif checkbit == '10':
                string += 'C'
            elif checkbit == '11':
                string += 'G'
            checkbit = ''
            counter = 0
    characterString.append(string)
    string = ''

print("\nafter convert to dna:\n")
counter = 0
for item in characterString:
    counter += 1
    print("item "+str(counter)+" : "+str(item) + " With size : "+str(len(item)))

output = open("output.txt", "w")
output.writelines(characterString)
output.close()
```
